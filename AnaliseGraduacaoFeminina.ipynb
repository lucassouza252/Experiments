## Análise de formações em cursos superiores realizados por mulheres entre os anos de 1970 a 2011 nos Estados Unidos

```python
#Importando bibliotecas para análise de dados
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sea
```

```python
#Lendo csv e salvando num DataFrame Pandas
df_bachelors = pd.read_csv('Percent-bachelors-degrees-women-usa.csv')
```

#### Verificação dos dados, suas informações e tipos de variáveis

```python
#Imprimindo as 5 primeiras linhas para visualizar o tipo de variavel
df_bachelors.head()
```

```python
#Fazendo com que 'Year' se torne o id, e removendo a coluna id do DataFrame
df_bachelors.index = df_bachelors['Year']
del df_bachelors['Year']
```

```python
#Verificando novamente as 5 primeiras colunas
df_bachelors.head()
```

```python
#Visualizando as informações estatisticas dos dados
df_bachelors.describe()
```

### Com base nos gráficos seguintes podemos perceber uma grande diminuição em alunas mulheres nos cursos de exatas e ciências da computação com o passar dos anos, em contra partida há um grande aumento nos cursos de biológicas e agrilcultura.

```python
#Medias da porcentagem de alunas por ano em seus cursos
for x in df_bachelors.columns:
    plt.scatter(df_bachelors.index, df_bachelors[x])
    plt.xlabel('Year')
    plt.ylabel('Average')
    plt.title(x)
    plt.show()
```

```python
#Busca as duas primeiras letras de cada nome de coluna para melhor visualização no grafico posterior
first_letters = []
for x in df_bachelors.columns:
    first_letters.append(x[0:2])
```

### Embora haja um grande aumento de alunas nos cursos de Biológicas e Agronômicas, a média anual é crescente em cursos da área de Saúde e Educação, sendo mais baixas nas áreas de Física e Agronômia.

```python
#Gráfico de média de alunas por curso por ano
plt.bar(first_letters, df_bachelors.mean())
plt.xlabel('Curses')
plt.ylabel('Average of Students')
plt.title('Average per curse')
plt.show()
```

```python
#Melhor visualização de Agricultura pois suas dispersão é grande devido ao seu aumento rápido
plt.bar(df_bachelors.index, df_bachelors.Agriculture)
plt.xlabel('Year')
plt.ylabel('Curse')
plt.title('Agriculture')
plt.show()
```

### Podemos notar que há 3 outliers nos dados de Agricultura, demonstrando assim sua diferença entre curva de crescimento e a média.

```python
#Utilizando biblioteca seaborn para melhor visualização do boxplot
sea.boxplot(df_bachelors.Agriculture)
plt.show()
```

```python
#Visualização da distribuição dos cursos por média
fatias = df_bachelors.mean()
nomes = df_bachelors.columns
plt.pie(fatias, labels = nomes, startangle = 90, shadow = True)
plt.show()
```
